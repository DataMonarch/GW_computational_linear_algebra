```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create two random vectors
v1 = np.random.randint(10, size=2)
v2 = np.random.randint(10, size=2)
```

```python
print(f'Vector 1: {v1}')
print(f'Vector 2: {v2}')

width = 0.01
head_width = 0.1
ax_lim = 12
# Calculate the sum of the two vectors
v_sum = v1 + v2

print(f'Sum: {v_sum}')

# Create a new figure
fig, ax = plt.subplots(figsize=(10, 10))

# set the x and y axis limits as 12
ax.set_xlim(0, ax_lim)
ax.set_ylim(0, ax_lim)

# Plot the first vector
ax.arrow(0, 0, v1[0], v1[1], color='blue', label='Vector 1', 
    width=width, head_width=head_width, length_includes_head=True)

# Plot the first vector starting at the end of the second vector as dotted line
ax.arrow(v2[0], v2[1], v1[0], v1[1], color='blue', linestyle='dashed',
    width=width, head_width=head_width, length_includes_head=True)

# Plot the second vector
ax.arrow(0, 0, v2[0], v2[1], color='orange', label='Vector 2', 
    width=width, head_width=head_width, length_includes_head=True)

# Plot the second vector starting at the end of the first vector as dotted line
ax.arrow(v1[0], v1[1], v2[0], v2[1], color='orange', linestyle='dashed',
    width=width, head_width=head_width, length_includes_head=True)

# Plot the sum of the two vectors
ax.arrow(0, 0, v_sum[0], v_sum[1], color='green', label='Sum', 
    width=width, head_width=head_width, length_includes_head=True)

# set the origin of the new coordinate system on the original coordinate system
origin_new_coord = np.array([1, 2]) # (x, y)
# compute the coordinates of the vectors in the new coordinate system
v1_new_coord = v1 - origin_new_coord
v2_new_coord = v2 - origin_new_coord
v_sum_new_coord = v_sum - origin_new_coord

print(f'Origin of new coordinate system: {origin_new_coord}')
print(f'Vector 1 in new coordinate system: {v1_new_coord}')
print(f'Vector 2 in new coordinate system: {v2_new_coord}')
print(f'Sum in new coordinate system: {v_sum_new_coord}')


# plot dashed lines representing the y axis of new coordinate system
ax.plot([origin_new_coord[0], origin_new_coord[0]], [origin_new_coord[1], origin_new_coord[1]+ax_lim], 
        color='black', linestyle='dashed', alpha=0.5)
# plot dashed lines representing the x axis of new coordinate system
ax.plot([origin_new_coord[0], origin_new_coord[0]+ax_lim], [origin_new_coord[1], origin_new_coord[1]],
        color='black', linestyle='dashed', alpha=0.5)

# define the labels of the vectors
v1_label = f'({v1[0]}, {v1[1]}) -> ({v1_new_coord[0]}, {v1_new_coord[1]})'
v2_label = f'({v2[0]}, {v2[1]}) -> ({v2_new_coord[0]}, {v2_new_coord[1]})'
v_sum_label = f'({v_sum[0]}, {v_sum[1]}) -> ({v_sum_new_coord[0]}, {v_sum_new_coord[1]})'

# Add labels to the vectors
ax.annotate(v1_label, (v1[0], v1[1]), xytext=(v1[0], v1[1]+1))
ax.annotate(v2_label, (v2[0], v2[1]), xytext=(v2[0], v2[1]+1))
ax.annotate(v_sum_label, (v_sum[0], v_sum[1]), xytext=(v_sum[0], v_sum[1]+1))


# # Add legend
ax.legend(loc='upper left')
# ax2.legend(loc='upper right')

# Show the plot
plt.show()
```
